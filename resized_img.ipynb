## CV2.resize with calculation

```python
import cv2
import matplotlib.pyplot as plt
```

```python
path = r'C:\Users\Clsm\all_ROI_CLEAN_train\sit_stand_dataset\sit\DS_006_002_003_001_000001.jpg'
img = cv2.imread(path)
plt.title("Original")
plt.imshow(img)
```

# let's resize our image to be 224 pixels wide, but in order to
# prevent our resized image from being skewed/distorted, we must
# first calculate the ratio of the *new* width to the *old* width

```python
r = 224.0/img.shape[1]
dim = (224, int(img.shape[0] * r))
dim
```

```python
resized = cv2.resize(img,dim,interpolation=cv2.INTER_AREA)
plt.title("Resized")
plt.imshow(resized)
```

## Get Color Code from Image

```python
from PIL import Image
import os
image_path = r'C:\Users\Clsm\Pictures\Screenshots\334493734.jpg'
```

```python
def hex_to_code(r,g,b):
    return '#{:02x}{:02x}{:02x}'.format(r,g,b)
```

```python
UniqueColor = set()
def get_Color_code(image_path):
    image = Image.open(image_path)
    pixels = image.load()
    for i in range(image.size[0]):
        for j in range(image.size[1]):
            r,g,b = pixels[i,j]
            hexColorCode = hex_to_code(r,g,b)
            UniqueColor.add(hexColorCode)
    return UniqueColor
```

```python
get_Color_code(image_path)
```
