```python
image = cv2.imread(r'C:\Users\Clsm\Downloads\1.jpeg')
gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
```

```python
gray_r = gray.reshape(gray.shape[0]*gray.shape[1])
for i in range(gray_r.shape[0]):
    if gray_r[i] > gray_r.mean():
        gray_r[i] = 1
    else:
        gray_r[i] = 0
gray = gray_r.reshape(gray.shape[0],gray.shape[1])
plt.imshow(gray, cmap='gray')
```

```python
path = r'C:\Users\Clsm\Downloads\1.jpeg'
image = cv2.imread(path)/255
plt.imshow(image)
print(image.shape)
```

```python
pic_n = image.reshape(image.shape[0]*image.shape[1], image.shape[2])
pic_n.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=0).fit(pic_n)
pic2show = kmeans.cluster_centers_[kmeans.labels_]
```

```python
cluster_pic = pic2show.reshape(image.shape[0], image.shape[1], image.shape[2])
plt.imshow(cluster_pic)
```
