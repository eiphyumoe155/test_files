```python
count = 0
for child in body.children:
    if child.name is not None:
        print(str(child.name) + ":" + str(type(child)), end = '\t')
        print(child.children)
        count = count + 1
print(count)
```

```python
temp = []
def walker(soup):
    if soup.name is not None:
        count  = 0
        for child in soup.children:
            #process node
            print(str(child.name) + ":" + str(type(child)))
            walker(child)
            count = count +1
        print("fininshed : ",count)
        temp.append(count)
```
