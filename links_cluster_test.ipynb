```python
import numpy as np
from copy import deepcopy
import numpy as np
from copy import deepcopy
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

```python
    
page_url = r"https://sakabanashi.takarashuzo.co.jp/cat2"
domain = r"https://sakabanashi.takarashuzo.co.jp"
```

```python
driver_path = r'chromedriver.exe'
driver = webdriver.Chrome(driver_path)
driver.get(page_url)
htmlContent = driver.page_source
soup = BeautifulSoup(htmlContent, 'html.parser')
soupBody = soup.body
```

```python
soupBody
```

```python
urls = soupBody.findAll('a')
dict_urls = set()
#print(len(urls))
for url in urls:
    url_href = url['href']
    if len(url_href)<10:
        continue
    dict_urls.add(url_href)
print(dict_urls)
    
```

```python
max_columns = 0
data_rows = []
for current_url in dict_urls:
    url = current_url.split('/')
    if len(url)-1 > max_columns:
        max_columns = len(url)-1
    data_rows.append(url[1:])
print(data_rows)
print(max_columns)
```

```python
new_data_rows = []
row_records = {}

for cur_row in data_rows:
    temp_cur_row = []
    for i in range(0,max_columns,1):
        row_col_data = ''
        try:
            row_col_data = cur_row[i]
        except IndexError:
            row_col_data = ''
        temp_cur_row.append(row_col_data)
    
        try:
            row_records[i].add(row_col_data)
            
        except KeyError:
            col_dist = set()
            col_dist.add(row_col_data)
            row_records[i] = col_dist
    new_data_rows.append(temp_cur_row)
print(new_data_rows)
#print(row_records)
```

```python
import numpy as np
index_new_data_table = []
new_temp_row = []
for row in new_data_rows:
    for i in range(0, len(row),1):
        dataMark_str = new_data_rows[i]
        for j in range(0, len(dataMark_str), 1):
            #print(dataMark_str[j])
            if row[i] == dataMark_str[j]:
                #print("Equal")
                new_temp_row.append(j)
                break
            else:
                continue
    index_new_data_table.append(new_temp_row)
print(index_new_data_table)
```
